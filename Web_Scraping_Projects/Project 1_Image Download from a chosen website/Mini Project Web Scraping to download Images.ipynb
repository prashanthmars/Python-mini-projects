```python
import requests
```

```python
res = requests.get("https://www.gettyimages.in/photos/crossfit?sort=mostpopular&mediatype=photography&phrase=crossfit")
```

```python
type(res)
```

```python
res.text
```

```python
import bs4
```

```python
soup = bs4.BeautifulSoup(res.text,"lxml")
soup
```

```python
rawtext=soup.select('img')
'src' in rawtext[0]
print('src' in str(rawtext[0]))
```

```python
imagelinks=[]
for eachimagelinkrow in rawtext:
    #print(eachimagelinkrow)
    #print('\n')
    if 'src' in str(eachimagelinkrow):
        imagelinks.append(eachimagelinkrow['src'])
    else:
        next
imagelinks
```

```python
imagelinks=list(filter(lambda x: 'gettyimages' in x,imagelinks))
```

```python

for i,link in enumerate(imagelinks):
    image_link = requests.get(link)
    f=open(f"downloaded images\\image{i}.jpg","wb")
    f.write(image_link.content)
    print(f"{i} out of {len(imagelinks)}")
    f.close()
```
